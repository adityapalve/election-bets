```python
#We now need to convert all the "favorites" into integers as some are integers and some strings. Pandas might automaticall do this for us
trump = trump.astype({'favorites': 'int'})
trump = trump.astype({'retweets': 'int'})
trump.info()
```

```python

# Here we have created a trump data set which displays all the deleted tweets. We do this by cross correlating trumps deleted tweets with the list of all his tweets
#The retweets have 0 favourites (which are basically likes) and have an RT at the front 
# To distinquish between the tweets trump deleted and the tweets deleted by twitter (restricted), we make the assumption that whenever a tweet is deleted by twitter the 
# like and retweet numbers are set to 0. We use this to distrinquish 

new_status = list()
for index,val in enumerate(trump['status']):
    list_fav = list(trump['favorites'])
    list_retweets = list(trump['retweets'])
    if(val=="Normal" and list_fav[index]==0 and list_retweets[index] == 0):
        new_status.insert(index, "Restricted")
    else:
        new_status.insert(index,val)
trump['status'] = new_status
trump[trump['status'] == "Deleted"]

            
```

```python
#removed the tweets with zero likes (restricted and retweeted) as these accounted only for aprx 50/1500 data points.
# We did this because we want to analyse the ratio of retweets to likes. Having 0 likes provides no information. 
biden_tweet_retweet_ratio = biden.copy()
intermediate = trump[trump['isRetweet'] == False]
trump_tweet_retweet_ratio = intermediate[intermediate["status"] != "Restricted"]
```

```python
#We now attempt to see about how the engagment with each candidate evolves over each time by looking at retweet to tweet like ratio
ratio_biden = list()
for i in range(len(biden_tweet_retweet_ratio['tweet_id'])):
    ratio_biden.insert(i,list(biden_tweet_retweet_ratio['tweet_retweet_count'])[i]/list(biden_tweet_retweet_ratio['tweet_like_count'])[i])
ratio_trump = list()
for i in range(len(trump_tweet_retweet_ratio['tweet_id'])):
    ratio_trump.insert(i,list(trump_tweet_retweet_ratio['tweet_retweet_count'])[i]/list(trump_tweet_retweet_ratio['tweet_like_count'])[i])

        
biden_tweet_retweet_ratio['retweet_tweet_like_ratio'] = ratio_biden
trump_tweet_retweet_ratio['retweet_tweet_like_ratio'] = ratio_trump
end = list(trump_tweet_retweet_ratio['tweet_day'])[0]
start = list(trump_tweet_retweet_ratio['tweet_day'])[-1]
increment = datetime.timedelta(days=1)
ratio_biden_average = list()
ratio_trump_average = list()
dates = trump_tweet_retweet_ratio['tweet_day'].unique()
index=0
while(start <= end):
    average_biden = biden_tweet_retweet_ratio.loc[(biden_tweet_retweet_ratio['tweet_day'] == start), 'retweet_tweet_like_ratio'].mean()
    average_trump = trump_tweet_retweet_ratio.loc[(trump_tweet_retweet_ratio['tweet_day'] == start), 'retweet_tweet_like_ratio'].mean()
    ratio_biden_average.insert(index,average_biden)
    ratio_trump_average.insert(index,average_trump)
    index+=1
    start += increment 
```

```python
plt.figure(2)
plt.plot(ratio_biden_average,label='Biden ratio')
plt.ylim(0, 0.3)
plt.plot(ratio_trump_average, label="Trump ratio")
plt.ylim(0, 0.3)
plt.xlabel("Days")
plt.ylabel("Ratio")
plt.title("Retweet to Like Ratio for Trump and Biden")
plt.legend()
plt.show()
```

```python
from collections import Counter
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
stopwords_dict = Counter(stop_words)
```

```python
def clean_tweets(tweets):
    cleaned_tweets = list()
    for index,tweet in enumerate(tweets):
        tweet = re.sub(r"|", "",tweet)
        tweet = ' '.join([word for word in tweet.split() if word not in stopwords_dict])
        cleaned_tweets.insert(index,tweet)
    return cleaned_tweets

cleaned_tweets_trump = clean_tweets(list(T_US_lan_time['clean_tweet']))
cleaned_tweets_biden = clean_tweets(list(B_US_lan_time['clean_tweet']))
T_US_lan_time['clean_tweet'] = cleaned_tweets_trump
B_US_lan_time['clean_tweet'] = cleaned_tweets_biden

```

```python
def convert_emoji_emojicon(sentence):
    sentence = sentence.replace("üôÇ",":)")
    sentence = sentence.replace("üòä",";)")
    sentence = sentence.replace("üòÄ",":D!")    
    sentence = sentence.replace("üòÅ",":D!") 
    sentence = sentence.replace("üòâ",";‚Äë)")    
    sentence = sentence.replace("üòÇ","XD")
    sentence = sentence.replace("üòÜ","XD")
    sentence = sentence.replace("üôÅ",":(")  
    sentence = sentence.replace("üò≠",";(!")
    sentence = sentence.replace("üòé","|;‚Äë)")
    sentence = sentence.replace("üò°",">:(!")
    sentence = sentence.replace("üò†",">:(!")
    sentence = sentence.replace("üòç",":√ó!")
    sentence = sentence.replace("ü§®","',:-|")
    return sentence
#Replace "converted_emoji_emojicon" with "clean_tweet" to remove emoji's
T_US_lan_time['converted_emoji_emojicon'] = T_US_lan_time['clean_tweet'].apply(lambda x : convert_emoji_emojicon(x))
B_US_lan_time['converted_emoji_emojicon'] = B_US_lan_time['clean_tweet'].apply(lambda x : convert_emoji_emojicon(x))
```

## Actual US election results
In vast majority of states, the winner of the popular vote takes all the electoral votes for that state. We therefore have summed county votes into states to map the actual election results. 

It is worth noting there are two exceptions to the above assumption: Maine and Nebraska Instead, they use a "district system," where two electoral votes are allocated to the state-wide winner, and one electoral vote is allocated to the winner in each of the state's congressional districts.

```python
df_votes_by_county = pd.read_csv('/kaggle/input/us-election-2020/president_county_candidate.csv')
df_votes_by_county

state_codes = pd.read_csv('/kaggle/input/latitude-and-longitude-for-every-country-and-state/world_country_and_usa_states_latitude_and_longitude_values.csv')
state_codes = state_codes[['usa_state','usa_state_code']]
df_votes_by_county = df_votes_by_county.merge(state_codes, left_on='state', right_on='usa_state')
```

```python
#group by state and party, summing the total_votes column
# filter the DataFrame to include only 'DEM' and 'REP' parties
df_DEMREP = df_votes_by_county.loc[df_votes_by_county['party'].isin(['DEM', 'REP'])]

# create a pivot table with 'state' as index, 'party' as columns, and 'total_votes' as values
df_DEMREP = pd.pivot_table(df_DEMREP, values='total_votes', index='state', columns='party', aggfunc=sum)
df_DEMREP = df_DEMREP.merge(state_codes, left_on='state', right_on='usa_state')
df_DEMREP['percent_JOE'] =  df_DEMREP['DEM'] * 100 / (df_DEMREP['DEM']+df_DEMREP['REP'])
df_DEMREP['percent_Trump'] =  df_DEMREP['REP'] * 100 / (df_DEMREP['DEM']+df_DEMREP['REP'])
df_DEMREP['softmax percentage'] = 100* np.exp(df_DEMREP['percent_JOE']/100)/(np.exp(df_DEMREP['percent_JOE']/100) + np.exp(df_DEMREP['percent_Trump']/100))
```

```python
fig = px.choropleth(df_DEMREP, locations='usa_state_code', locationmode='USA-states',
                    color='percent_JOE', scope='usa', color_continuous_scale = 'RdBu',  range_color=[25,75],
                    title='Popular vote per state in US election')
fig.show()
```

#### Model prediction errors
We measure the error in both our sentiment models, by their accuracy in aligning with results of the election. 

    Note: in this we assume that 0.5 sentiment demonstrates neutral stance between candidates  >0.5 = Joe favoured, <0.5 = Trump favoured

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

#VADER ERROR
#Make new dataframe of textblob ratios and actual election results
vader_vs_actual = intersection_df = pd.merge(df_DEMREP[['usa_state_code', 'percent_JOE','softmax percentage']], 
                                                Joined_truncated_df[['state_code', 'ratio']] , 
                                                left_on='usa_state_code', right_on='state_code')
vader_vs_actual = vader_vs_actual.drop('state_code', axis=1)
vader_vs_actual['ratio'] = vader_vs_actual['ratio']*100 #Convert to percent to compare with reality
vader_vs_actual

#Calculate mean absolute error
mae_textblob = mean_absolute_error(vader_vs_actual['percent_JOE'], vader_vs_actual['ratio'])
print('Vader  MAE: ', round(mae_textblob,3), '%')

#Calculate root mean squared error
rmse_textblob = np.sqrt(mean_squared_error(vader_vs_actual['percent_JOE'], vader_vs_actual['ratio']))
print('Vader  RMSE: ', round(rmse_textblob,3), '%')

#calcuate softmax error 
softmax_textblob = mean_absolute_error(vader_vs_actual['softmax percentage'], vader_vs_actual['ratio'])
print('Vader softmax error: ', round(softmax_textblob,3), '%')
```

#### Penalising incorrect prediction errors
The error values look quite low, but the maps look quite different. This is because whulst the magnitude of error is small, it crosses the 50% boundary and hence predicts the wrong candidate. 
We therefore compute a modified error measure, where a higher weighting is put on errors that results in the wrong prediction.

```python
# Compute the modified MAE that places a higher penalty on errors on the opposite side of the threshold (resulting in wrong candidate prediction)
thresh = 50 
abs_error = np.abs(vader_vs_actual['percent_JOE'] - vader_vs_actual['ratio'])
modified_mae = np.mean(np.where((vader_vs_actual['ratio'] - thresh) * (vader_vs_actual['percent_JOE'] - thresh) < 0, 2 * abs_error, abs_error))

# Print the modified MAE
print("Boundary-weighted Vader MAE:", round(modified_mae,2),'%')
```

```python
# Convert the predicted sentiment ratios to binary values
predicted_binary = np.where(vader_vs_actual['ratio'] > thresh, 1, 0)
# Convert the actual sentiment ratios to binary values
actual_binary = np.where(vader_vs_actual['percent_JOE'] > thresh, 1, 0)
# Calculate the absolute difference between the predicted binary values and actual binary values
abs_diff = np.abs(predicted_binary - actual_binary)

# Count the number of states where the absolute difference is 1
wrong_side_count = np.sum(abs_diff)
print("No. of incorrectly predicted states: ",wrong_side_count)
```
